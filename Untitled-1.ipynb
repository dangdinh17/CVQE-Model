```python
import numpy as np
import cv2
import os
import re  # để sử dụng biểu thức chính quy

def parse_yuv_filename(filename):
    # Ví dụ: BasketballDrill_832x480_500.yuv
    match = re.search(r'_(\d+)x(\d+)_(\d+)\.yuv$', filename)
    if match:
        width = int(match.group(1))
        height = int(match.group(2))
        num_frames = int(match.group(3))
        return width, height, num_frames
    else:
        raise ValueError(f"Filename format not recognized: {filename}")

def read_yuv420_frame(file_path, width, height):
    frame_size = width * height * 3 // 2  # YUV420
    with open(file_path, 'rb') as f:
        raw = f.read(frame_size)
        if len(raw) < frame_size:
            raise ValueError("File too short for a single frame.")
        y = np.frombuffer(raw[0:width*height], dtype=np.uint8).reshape((height, width))
        u = np.frombuffer(raw[width*height:width*height + (width//2)*(height//2)], dtype=np.uint8).reshape((height//2, width//2))
        v = np.frombuffer(raw[width*height + (width//2)*(height//2):], dtype=np.uint8).reshape((height//2, width//2))
        # Upsample U and V to match Y size
        u_up = cv2.resize(u, (width, height), interpolation=cv2.INTER_LINEAR)
        v_up = cv2.resize(v, (width, height), interpolation=cv2.INTER_LINEAR)
        yuv = cv2.merge([y, u_up, v_up])
        bgr = cv2.cvtColor(yuv, cv2.COLOR_YUV2BGR)
        return bgr

# Folder chứa các file .yuv
folder_path = r'MFQEv2-TPAMI\test_18\HM16.5_LDP\QP37'
output = r'output_frame0'
for filename in os.listdir(folder_path):
    if filename.endswith('.yuv'):
        width, height, num_frames = parse_yuv_filename(filename)
        print(f"{filename}: width={width}, height={height}, frames={num_frames}")
        
        yuv_path = os.path.join(folder_path, filename)
        output_path = os.path.join(output, f'{filename[:-4]}_frame0_compressed.png')
        
        frame = read_yuv420_frame(yuv_path, width, height)
        cv2.imwrite(output_path, frame)
        print(f"Saved: {output_path}")
```

```python
import re

def wrap_title_with_double_braces(bib_file_path):
    with open(bib_file_path, 'r', encoding='utf-8') as file:
        content = file.read()

    # Regex để tìm phần title={...}
    pattern = r'title\s*=\s*\{(.*?)\}'
    
    def replacer(match):
        original_title = match.group(1)
        # Tránh bọc lại nếu đã có {}
        if original_title.startswith('{') and original_title.endswith('}'):
            return f"title={{{original_title}}}"
        return f"title={{{{{original_title}}}}}"

    updated_content = re.sub(pattern, replacer, content, flags=re.DOTALL)

    # Ghi lại vào file mới hoặc đè lên file cũ
    with open('output.bib', 'w', encoding='utf-8') as file:
        file.write(updated_content)

# Gọi hàm với đường dẫn file .bib của bạn
wrap_title_with_double_braces('paper.bib')
```

```python
import re

def remove_url_field(bib_file_path):
    with open(bib_file_path, 'r', encoding='utf-8') as file:
        content = file.read()

    # Xoá dòng chứa 'url = {...}' hoặc 'url= {...}' hoặc tương đương, kể cả có hoặc không dấu phẩy
    updated_content = re.sub(r'^\s*url\s*=\s*\{.*?\},?\s*\n?', '', content, flags=re.IGNORECASE | re.MULTILINE)

    with open('output_no_url.bib', 'w', encoding='utf-8') as file:
        file.write(updated_content)

# Gọi hàm
remove_url_field('output.bib')
```

```python
import re

# Từ điển viết tắt cho journal và conference ngắn gọn
abbreviations = {
    "International Conference on Computer Vision": "ICCV",
    "Conference on Computer Vision and Pattern Recognition \\(CVPR\\)": "CVPR",
    "Conference on Computer Vision and Pattern Recognition": "CVPR",
    "European Conference on Computer Vision": "ECCV",
    "Transactions on Image Processing": "TIP",
    "Transactions on Pattern Analysis and Machine Intelligence": "TPAMI"
}

def abbreviate_bib_labels(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as f:
        bib_content = f.read()

    for long, short in abbreviations.items():
        # Gói cụm từ trong dấu ngoặc {} nếu có
        pattern = re.compile(r'({\s*' + long + r'\s*})')
        replacement = '{' + short + '}'
        bib_content = pattern.sub(replacement, bib_content)

        # Thử thay không có dấu ngoặc nếu có
        bib_content = re.sub(r'(?<!{)' + long + r'(?!})', short, bib_content)

    with open(output_file, 'w', encoding='utf-8') as f:
        f.write(bib_content)

    print(f"✅ Đã tạo file rút gọn tại: {output_file}")

# Sử dụng: đổi tên file nếu cần
abbreviate_bib_labels('output_abbreviated.bib', 'output_abbreviated.bib')
```

```python
import numpy as np
from scipy.interpolate import PchipInterpolator
from scipy.integrate import quad

def bd_rate(R1, PSNR1, R2, PSNR2):
    log_R1 = np.log(R1)
    log_R2 = np.log(R2)

    interp1 = PchipInterpolator(PSNR1, log_R1)
    interp2 = PchipInterpolator(PSNR2, log_R2)

    min_int = max(min(PSNR1), min(PSNR2))
    max_int = min(max(PSNR1), max(PSNR2))

    integral1, _ = quad(interp1, min_int, max_int)
    integral2, _ = quad(interp2, min_int, max_int)

    avg_diff = (integral2 - integral1) / (max_int - min_int)
    return (np.exp(avg_diff) - 1) * 100  # kết quả là %

# Ví dụ dùng:
R1 = [500, 1000, 1500, 2000]
PSNR1 = [30.1, 33.4, 35.0, 36.1]

R2 = [500, 1000, 1500, 2000]
PSNR2 = [29.8, 32.9, 34.4, 35.6]

print(f"BD-Rate: {bd_rate(R1, PSNR1, R2, PSNR2):.2f}%")
```

```python
import ffmpeg
import os

def encode_to_h264(input_path, output_path, codec='libx264', format='yuv420p', width=1920, height=1080):
    """
    Encode video từ file YUV hoặc MP4 sang H.264/H.265.
    
    Args:
        input_path (str): Đường dẫn file đầu vào (MP4 hoặc YUV).
        output_path (str): Đường dẫn file đầu ra (H.264/H.265).
        codec (str): 'libx264' (H.264) hoặc 'libx265' (H.265).
        format (str): Định dạng pixel (yuv420p, nv12, ...).
        width (int): Chiều rộng video (nếu đầu vào là YUV).
        height (int): Chiều cao video (nếu đầu vào là YUV).
    """
    try:
        if input_path.endswith('.yuv'):
            # Nếu đầu vào là file YUV (cần chỉ định width, height, format)
            (
                ffmpeg
                .input(input_path, format='rawvideo', pix_fmt=format, s=f'{width}x{height}')
                .output(output_path, vcodec=codec, pix_fmt=format)
                .overwrite_output()
                .run()
            )
        else:
            # Nếu đầu vào là MP4/MKV/AVI...
            (
                ffmpeg
                .input(input_path)
                .output(output_path, vcodec=codec, pix_fmt=format)
                .overwrite_output()
                .run()
            )
        print(f"✅ Encoded thành công: {output_path}")
    except ffmpeg.Error as e:
        print(f"❌ Lỗi encode: {e.stderr.decode()}")

# Ví dụ:
encode_to_h264(
    input_path='output\BasketballDrive_1920x1080_500.yuv',  # Hoặc 'input.yuv'
    output_path='output\BasketballDrive_1920x1080_500_encoded.mp4',
    codec='libx264',  # Thay 'libx265' để encode H.265
    format='yuv420p',
    width=1920,  # Bỏ qua nếu đầu vào là MP4
    height=1080  # Bỏ qua nếu đầu vào là MP4
)
```
